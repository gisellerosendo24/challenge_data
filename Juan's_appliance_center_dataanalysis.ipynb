### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.info()
```

```python
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'
df_total = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

import matplotlib.pyplot as plt
import pandas as pd

# Combina tus tiendas si no lo has hecho aún
df_total = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

# Asegúrate de que cada tienda tenga su nombre en la columna 'Tienda'
df_total['Tienda'] = df_total['Tienda'].astype(str)

# Agrupa y reestructura
data_grouped = df_total.groupby(['Tienda', 'Categoría del Producto'])['Cantidad de cuotas'].sum().unstack(fill_value=0)

# Crear lista única de categorías
categorias = data_grouped.columns.tolist()

# Generar colores únicos por categoría (puedes personalizar estos colores)
import matplotlib.cm as cm
import numpy as np

cmap = cm.get_cmap('tab10', len(categorias))  # o 'tab20', 'Set3', etc.
category_colors = {cat: cmap(i) for i, cat in enumerate(categorias)}

# Crear gráfico
fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharex=True)
axes = axes.flatten()
tiendas = data_grouped.index

for i, tienda in enumerate(tiendas):
    ax = axes[i]
    ventas = data_grouped.loc[tienda]

    # Ordenar por ventas si deseas
    ventas = ventas.sort_values()

    # Obtener colores correspondientes
    colors = [category_colors[cat] for cat in ventas.index]

    # Graficar
    ax.barh(ventas.index, ventas.values, color=colors)
    ax.set_title(f'Tienda: {tienda}')
    ax.set_xlabel('Numero de ventas')
    ax.invert_yaxis()  # Categoría más vendida arriba

handles = [plt.Rectangle((0, 0), 1, 1, color=category_colors[cat]) for cat in categorias]
fig.legend(handles, categorias, title='Categorías', bbox_to_anchor=(1.05, 0.5), loc='center left')

plt.tight_layout()
plt.suptitle('Ventas por categoría', fontsize=16, y=1.03)
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Supongamos que ya tienes estas Series: max1, min1, etc.

# Crear el DataFrame con nombres y cantidades
df = pd.DataFrame([
    ['Tienda 1', max1.index[0], max1.values[0], min1.index[0], min1.values[0]],
    ['Tienda 2', max2.index[0], max2.values[0], min2.index[0], min2.values[0]],
    ['Tienda 3', max3.index[0], max3.values[0], min3.index[0], min3.values[0]],
    ['Tienda 4', max4.index[0], max4.values[0], min4.index[0], min4.values[0]]
], columns=['Tienda', 'Producto mas', 'Producto mas vendido', 'Producto menos', 'Producto menos vendido'])

# Preparar para graficar
df_plot = df[['Tienda', 'Producto mas vendido', 'Producto menos vendido']]
df_plot.set_index('Tienda', inplace=True)

# Crear gráfico
ax = df_plot.plot(kind='bar',
                  figsize=(10, 6),
                  color=['green', 'red'],
                  title='Producto más y menos vendido por tienda')

plt.ylabel('Cantidad de ventas')
plt.xticks(rotation=0)

# Añadir etiquetas con los nombres de productos sobre las barras
for idx, row in df.iterrows():
    ax.text(x=idx - 0.18,  # posición horizontal (ligeramente a la izquierda)
            y=row['Producto mas vendido'] + 1,  # ligeramente por encima
            s=row['Producto mas'],
            color='black',
            fontsize=9,
            rotation=360)

    ax.text(x=idx + 0.05,  # posición horizontal (ligeramente a la derecha)
            y=row['Producto menos vendido'] + 1,
            s=row['Producto menos'],
            color='black',
            fontsize=9,
            rotation=360)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Asegúrate de que cada DataFrame tenga columna 'Tienda'
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Unir todos
df_total = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

plt.figure(figsize=(10, 8))
for tienda, color in zip(df_total['Tienda'].unique(), ['Reds', 'Blues', 'Greens', 'Purples']):
    data = df_total[df_total['Tienda'] == tienda]
    sns.kdeplot(
        x=data['lon'],
        y=data['lat'],
        fill=True,
        cmap=color,
        alpha=0.5,
        label=tienda
    )

plt.title('Heatmap de densidad de ventas por tienda')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend()
plt.tight_layout()
plt.show()
```
